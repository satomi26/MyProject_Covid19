I want to see the correlation between median age and deaths number(deaths per milion) 
Let's try to make series now, country name and its median age.

```python
med_age = df.groupby("location")["median_age"].mean().sort_values(ascending=False)
```

```python
med_age.unique
```

```python
med_age1 = df.groupby("location")["median_age"].mean().sort_values(ascending=False)[:20]
plt.figure(figsize=(20,5))
sns.barplot(x = med_age1.index,y = med_age1.values, alpha=0.8)
plt.title("population by continent")
plt.ylabel('Number of Death', fontsize=12)
plt.xlabel('Country/Region', fontsize=12)
plt.show()
```

I want to see the correlation between median age and deaths number(deaths per milion) 
Let's try to make series, country name and its new_death_per_million.

```python
deaths_mill = df.groupby("location")["new_deaths_per_million"].sum().sort_values(ascending=True)
```

```python
deaths_mill = deaths_mill.round(0).astype(int)
```

```python
deaths_mill = deaths_mill[deaths_mill!=0]
```

```python
d = deaths_mill.sort_values(ascending=False)[:20]
plt.figure(figsize=(20,5))
sns.barplot(x = d.index, y = d.values)
plt.title("Deaths per Million due to covid-19")
plt.ylabel('Deaths per Million', fontsize=12)
plt.xlabel('Country/Region', fontsize=12)
plt.show()
```

```python
d1 = deaths_mill.sort_values(ascending=True)[:20]
plt.figure(figsize=(20,5))
sns.barplot(x = d1.index, y = d1.values)
plt.title("Deaths per Million due to covid-19")
plt.ylabel('Deaths per Million', fontsize=12)
plt.xlabel('Country/Region', fontsize=12)
plt.show()
```

Now I want to combine two series, deats_mill and med_age

```python
correlation = pd.concat([ med_age,deaths_mill], axis = 1)
correlation
```
